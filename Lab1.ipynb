# **Визуализация распределения переменных**
**Описание:**
Создание визуализации для понимания распределения числовых и категориальных переменных.

1. **Гистограмма для числовой переменной `Bmi`:**

```python
sns.histplot(df['bmi'], kde=True)
plt.title('Распределение Bmi')
plt.xlabel('Bmi')
plt.ylabel('Частота')
plt.show()
```

2. **Гистограмма для числовой переменной `Age`:**

```python
sns.histplot(df['age'], kde=True)
plt.title('Распределение Age')
plt.xlabel('Age')
plt.ylabel('Частота')
plt.show()
```

**3. Столбчатая диаграмма для категориальной переменной `Class`:**

```python
sns.countplot(x='smoker', data=df)
plt.title('Распределение smoker')
plt.xlabel('smoker')
plt.ylabel('Количество людей')
plt.show()
```

**4. Столбчатая диаграмма для категориальной переменной `Class`:**

```python
sns.countplot(x='region', data=df)
plt.title('Распределение region')
plt.xlabel('region')
plt.ylabel('Количество людей')
plt.show()
```

**5. Столбчатая диаграмма для категориальной переменной `Class`:**

```python
sns.countplot(x='sex', data=df)
plt.title('Распределение sex')
plt.xlabel('sex')
plt.ylabel('Количество людей')
plt.show()
```

**6. Боксплот для числовой переменной `children`:**

```python
sns.boxplot(x=df['children'])
plt.title('Boxplot children')
plt.xlabel('children')
plt.show()
```

**7. Боксплот для числовой переменной `charges`:**

```python
sns.boxplot(x=df['charges'])
plt.title('Boxplot charges')
plt.xlabel('charges')
plt.show()
```
